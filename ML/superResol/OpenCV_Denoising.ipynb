```python
img = cv2.imread("original.jpg")
```

This algorithm contains two popular functions: 
fastNlMeansDenoising 
and fastNlMeansDenoisingColored. 

The first one is for grayscale images and the second one is for colored photos. In our case, we will use the one for colored images.

```python
denoise_1 = cv2.fastNlMeansDenoisingColored(img,None,3,3,7,21) 
denoise_2 = cv2.fastNlMeansDenoisingColored(img,None,5,5,7,21) 
denoise_3 = cv2.fastNlMeansDenoisingColored(img,None,15,15,7,21)
```

```python
cv2.imwrite('image_1.png', denoise_1) 
cv2.imwrite('image_2.png', denoise_2) 
cv2.imwrite('image_3.png', denoise_3)

cv2_imshow(cv2.imread('/content/image_1.png')) 
cv2_imshow(cv2.imread('/content/image_2.png')) 
cv2_imshow(cv2.imread('/content/image_3.png')) 
```
